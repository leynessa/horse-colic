#**Preprocessing**

---




Preprocessing is a crucial step in preparing the data for machine learning or data analysis tasks. It involves various data cleaning, transformation, and normalization techniques to ensure that the data is in a suitable format for modeling. Preprocessing can vary depending on the type of data and the specific requirements of the task, but some common preprocessing steps include:

1. **Handling Missing Data**: Dealing with missing values, which may involve imputation (replacing missing values with estimated values) or removing instances or features with too many missing values.

2. **Encoding Categorical Variables**: Converting categorical variables into a numerical format that can be understood by machine learning algorithms. This can be done using techniques like one-hot encoding or label encoding.

3. **Feature Scaling**: Scaling numerical features to a similar range to prevent certain features from dominating the learning process. Common scaling methods include normalization (scaling to [0, 1]) and standardization (scaling to mean 0 and standard deviation 1).

4. **Data Normalization**: Normalizing data to have a common scale, especially when features have different units or scales. It ensures that all features contribute equally to the model's learning process.

5. **Data Transformation**: Applying mathematical transformations to the data to make it more amenable to modeling. Examples include logarithmic transformations, square roots, or Box-Cox transformations for non-linear relationships.

6. **Feature Selection**: Selecting relevant features that contribute most to the predictive power of the model. Removing irrelevant or redundant features can improve the model's efficiency and reduce overfitting.

7. **Handling Imbalanced Data**: Addressing class imbalance in classification tasks by using techniques like oversampling, undersampling, or using class weights to give more importance to the minority class.


These are just some common preprocessing steps, and the exact techniques and order of applying them can vary based on the specific dataset and the machine learning model being used. Preprocessing is essential to ensure that the data is in an appropriate format and quality for the model to learn effectively and make accurate predictions.

##**Removing the unwanted columns**

The following columns will be removed:
1. the columns "nasogastric_reflux_ph", "abdomo_protein" and "abdomo_appearance " are the top 3 features with the most missing values<br>
2. The columns "hospital_number" and "cp_data" are not relevant to predict the outcomes <br>
3. 'lesion_1', 'lesion_2', and 'lesion_3' columns seems redudant because all lesion cases was operated on or autopsied, the surgical lesion will be kept as it gives us an idea of lesions

```python
cols_drop = ['nasogastric_reflux_ph', 'abdomo_appearance',
              'abdomo_protein', 'hospital_number', 'cp_data',
              'lesion_1', 'lesion_2', 'lesion_3' ]
df_clean= df_train.drop(cols_drop, axis=1)
dftest_clean= df_test.drop(cols_drop, axis=1)
```

```python
sns.heatmap(df_clean.isnull())
```

```python
sns.heatmap(dftest_clean.isnull())
```

```python
# Fill in missing numeric values with the mean value
cols_2 = ['rectal_temp', 'pulse', 'respiratory_rate', 'packed_cell_volume', 'total_protein']
#training data
for col in cols_2:
    df_clean[col].fillna(df_clean[col].mean(), inplace = True)
#test data
for col in cols_2:
    dftest_clean[col].fillna(dftest_clean[col].mean(), inplace = True)
```

```python
sns.heatmap(df_clean.isnull())
```

##Encoding Categorical Variables:

```python
# One-hot encode
cols = ['surgery', 'age', 'temp_of_extremities', 'peripheral_pulse', 'mucous_membrane',
                  'capillary_refill_time', 'peristalsis', 'abdominal_distention',
                  'nasogastric_tube', 'nasogastric_reflux', 'rectal_exam_feces', 'abdomen',
                  'pain', 'surgical_lesion']



for col in cols:
    df_clean = pd.concat([df_clean, pd.get_dummies(df_clean[col], prefix=col)], axis=1)
    df_clean.drop([col], axis=1, inplace=True)

# One-hot encode columns in dftest_clean
for col2 in cols:
    dftest_clean = pd.concat([dftest_clean, pd.get_dummies(dftest_clean[col2], prefix=col2)], axis=1)
    dftest_clean.drop([col2], axis=1, inplace=True)

#for col in cols:
 # df_clean= pd.concat([df_clean, pd.get_dummies(df_clean[col],prefix=col)], axis=1).drop([col],axis=1)


#for col2 in cols:
 #  dftest_clean= pd.concat([dftest_clean, pd.get_dummies(dftest_clean[col2],prefix=col)], axis=1).drop([col2],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
#  Instantiate the StandardScaler

scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.linear_model import LogisticRegression
from sklearn import tree
```

```python
dt_model = DecisionTreeClassifier()

# Step 6: Train the model using the scaled training data
dt_model.fit(X_train_scaled, y_train)

# Step 7: Make predictions on the scaled test data
y_pred = dt_model.predict(X_test_scaled)

# Step 8: Evaluate the model's performance on the test data
accuracy_dt = accuracy_score(y_test, y_pred)
print("Decision tree Accuracy:", accuracy_dt)

precision_dt = precision_score(y_test, y_pred, average=None)
print("Decision tree Precision:", precision_dt)

recall_dt = recall_score(y_test, y_pred, average=None)
print("Decision tree Recall:", recall_dt)
```

```python
## choose the correct max depth on the decision tree
```

```python
from sklearn.model_selection import GridSearchCV

# Assume X_train, y_train are the training data and labels
# Define the parameter grid with max depth values to search through
param_grid = {'max_depth': [None , 4, 5, 10, 15, 20]}
grid_search = GridSearchCV(dt_model, param_grid, cv=5)
grid_search.fit(X_train_scaled, y_train)
# Get the best max depth from the search results
best_max_depth = grid_search.best_params_['max_depth']
print("Best Max Depth:", best_max_depth)
```
